## Cleaning Data

We need to fill out (impute) the missing data for **Age** column. We can use mean of Age but it might not give the best values 
because they will be randomly assigned. Instead we will use the **Pclass** column to determine the age of the passengers.

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='Pclass', y='Age', data= train)
```

For the 1st class the median age is 37.

For the 2nd class the median age is 29.

For the 3rd class the median age is 24.

People(in the 1st class) take time to accumulate wealth and so it is evident that the IQR is between 28 and 49. On the other hand, younger people who do not have much wealth tend to buy cheaper ticets (3rd class).

```python
#Function to impute Age base on Pclass

def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
        
    else:
        return Age
```

```python
train['Age'] = train[['Age', 'Pclass']].apply(impute_age, axis =1)
```

```python
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap = 'viridis')
```

**We have imputed missing Age values**

**The Cabin columns contains too many missing values so we will drop it.**

```python
train.drop('Cabin', axis =1 , inplace= True)
```

```python
train.head()
```

```python
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap = 'viridis')
```

**Now we only need to deal with the Embarked column that only has 1 or 2 missing values. We will drop those missing values.**

```python
train.dropna(inplace = True)
```

```python
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap = 'viridis')
```

**No more missing values**

**Now we need to deal with Categorical Variables. We will convert them to dummy variables so our ML algo will be to interpret the numerical values**

```python
sex = pd.get_dummies(train['Sex'], drop_first=True)  #To avoid the issue of multi-collinearity
```

```python
embark = pd.get_dummies(train['Embarked'], drop_first= True)
```

```python
train = pd.concat([train, sex, embark], axis=1)
```

```python
train.head()
```

**We have the dummy variables concatenated so we can now drop the original columns and also few difficult to feature engineer column**

```python
train.drop(['Sex', 'Embarked', 'Name', 'Ticket'], axis= 1, inplace=True)
```

```python
train.head()
```

```python
train.drop('PassengerId', axis =1, inplace= True)
```

```python
train.head()
```

```python
X = train.drop('Survived', axis= 1)
```

```python
y = train['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state= 101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
y_pred = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_pred, y_test))
```
