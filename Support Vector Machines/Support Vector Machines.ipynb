# Support Vector Machines
It is a Supervised Learning Algorithm that can be used with Linear Regression and Classfication. SVM tries to find a margin that separates classes. It is usually used for classification. 

**hyperplane**: The goal of SVM is to find a hyperplane that would leave the widest possible "cushion" between input points from two classes. There is a tradeoff between "narrow cushion, little / no mistakes" and "wide cushion, quite a few mistakes". hyperplane is nothing but a classification boundary.

**kernel** : The kernel can be interpreted as corresponding to a feature transformation onto an infinite dimensional space. kernel parameter typically has following values:

 *linear*  (used for linear hyperplane)
 
 *rbf*     (used for non-linear hyperplane)
 
 *poly*    (used for non-linear hyperplane)

**C** is the cost of misclassification on the training data or can be termed as the 'penalty'

Large C makes the cost of misclassification high ('hard margin"), thus forcing the algorithm to explain the input data stricter and potentially overfit. Large C gives low bias and high variance

Small C makes the cost of misclassificaiton low ("soft margin"), thus allowing more of them for the sake of wider "cushion". Small C gives high bias and low variance

**gamma** is a parameter of Gaussian kernel to handle non-linear classification.

Some points are not linearly separable in 2D so you want to transform them to a higher dimension where they will be linearly separable. To "raise" the points you use the RBF kernel, gamma controls the shape of the "peaks" where you raise the points. A small gamma gives you a pointed bump in the higher dimensions, a large gamma gives you a softer, broader bump.

Large gamma will give you high bias and low variance.

Small gamma will give you low bias and high variance

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn import datasets
```

```python
iris_data = datasets.load_iris()
```

```python
X = iris_data.data[:, :2]
y = iris_data.target
```

### kernel = 'linear', C = 1.0, gamma= 'auto'

```python
svc = SVC(kernel= 'linear', C = 1.0, gamma= 'auto').fit(X,y)
```

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

h = (x_max / x_min)/100

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with linear kernel')
plt.show()
```

**soft margin: ** SVM that allows some examples to be "ignored" or placed on the wrong side of the margin. Soft margin leads to a better overall fit

**hard margin: ** Overfits the data. Outliers well within the boundaries can influence the margin.


**Advantages:**
1. Using regularization parameters, one can avoid the problem of overfitting. 

2. The kernel trick is useful in variety of cases. 

**Disadvantages:**
1. The biggest disadvantage is choosing appropriate hyperparameters (C and gamma) of the SVM that will allow for sufficient generalization performance.

2. Choosing the appropriate kernel function can be tricky.

3. On perceptual tasks (vision, speech etc), they are massively outclassed by deep neural networks. 

The goal is to find the balance between "not too strict" and "not too loose". Cross-validation and resampling, along with grid search, are good ways to finding the best C and Gamma hyper-parameters
